```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
from scipy.optimize import curve_fit
```

## Finding the Lyapunov Exponent for a Delta

This notebook finds the divergence between two nearby trajectories over time and fits an exponential curve.

## Prerequisites:

We need  base \(eta\_base\) and perturbed \(eta\_perturbed\) trajectories from chaos\_runs.m in the directory of this notebook. chaos\_runs.m needs the chaotic_base workspace for setting parameters for a chaotic delta.


## Demonstrating chaos

- Parameters match those in appendix table for a chaotic delta
- White noise is applied at all cells at each time step`perturbtype=1;`
- with amplitude $2*10^{-4}$
  `perturb_amp=2*10^-4`
- Some amount of white noise is necessary to explore dynamics near the unstable fixed point.



```python
eta_base = scipy.io.loadmat('eta_base.mat')['eta_save']
eta_base.shape
```

```python
eta_comp = scipy.io.loadmat('eta_perturbed.mat')['eta_save']
eta_comp.shape
```

Finding $|d|$ time series


```python
d = []
for t in range(eta_comp.shape[1]):
    d_j = 0
    for j in range(6):
        d_j += np.abs(eta_comp[j+1][t] - eta_base[j+1][t])
    d.append(d_j)

d = np.array(d)
```

```python
times = scipy.io.loadmat('eta_base.mat')['t_save'][0]
times.shape
```

```python
plt.plot(times, d)
```

## Curve Fitting



```python
popt, pcov = curve_fit(lambda t, b: d[0] * np.exp(b * t), times, d)
print(popt, pcov)
```

```python
plt.figure()
plt.plot(times, d)
plt.plot(times, d[0]*np.exp(popt[0]*times))
plt.xlabel('t')
plt.ylabel('$|d|$')
plt.legend(['$|d|$', '$y = d_0 e^{\lambda t}$, $\lambda = $' + str(np.round(popt, 2))])
```
